```python
import numpy as np

import matplotlib.pyplot as plt



n = 256

X = np.linspace(-np.pi, np.pi, n, endpoint=True)

Y = np.sin(2 * X)



plt.axes([0.025, 0.025, 0.95, 0.95])



plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.fill_between(X, 1, Y + 1, color='blue', alpha=.25) 
#Los primeros 3 parametros de la funcion fill_between son: x1,y1,y2 
#y basicamente lo que hace la funcion es llenar todo lo que hay entre y1 y y2
#x es igual a 256 numeros entre -3.14 a 3.14
#Como y1 siempre es 1 entonces (x,y1) seria (-3.14,1),(-3.12,1),(-3.10,1),... osea una linea recta en ordenada 1
#Y como y2 es igual a la grafica del seno+1 entonces junto con x hara la grafica del seno + 1
#Asi que ese espacio entre la linea recta de (x,y1) y las ondulaciones de (x,y2) sera llenado por la funcion



plt.plot(X, Y - 1, color='blue', alpha=1.00)

plt.fill_between(X, -1, Y - 1, (Y - 1) > -1, color='blue', alpha=.25)
plt.fill_between(X, -1, Y - 1, (Y - 1) < -1, color='red',  alpha=.25)
#Esto es igual a la grafica anterior pero tiene un parametro extra: Where
#Este parametro indica que solo actives la funcion cuando se cumpla la condicion
#Aqui el le esta diciendo que solo pinte de azul cuando y2 sean mayor a 1 y rojo cuando y2 sean menor a 1



#plt.xlim(-np.pi, np.pi) # Esto es solo para eliminar ejes, es solo decorativo
#plt.xticks(()) # Esto es solo para eliminar ejes, es solo decorativo
#plt.ylim(-2.5, 2.5) # Esto es solo para eliminar ejes, es solo decorativo
#plt.yticks(()) # Esto es solo para eliminar ejes, es solo decorativo



plt.show()
```

## 2. A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 2 en el enunciado.

```python
import numpy as np

import matplotlib.pyplot as plt



n = 1024

X = np.random.normal(0, 1, n) #Esto genera un numero random entre -3 y 3

Y = np.random.normal(0, 1, n) #Esto genera un numero random entre -3 y 3
#El primer parametro es el 'medio' o al menos la probabilidad mas alta
#El segundo para metro es que tanto se dispersa, en este ejemplo es de 1 en 1
#El tercer parametro es la cantidad de elementos random a generar


T = np.arctan2(Y, X) 
#Por alguna razon el numero que esto arroja Python lo interpreta como un color
#Basicamente para resolver esto es solo tomar el valor de T y usarlo como un color en plt.scatter

#plt.axes([0.025, 0.025, 0.95, 0.95]) #Esto es decorativo

plt.scatter(X, Y, c=T, alpha=.5)

#plt.xlim(-1.5, 1.5) #Esto es decorativo
#plt.xticks(()) #Esto es decorativo
#plt.ylim(-1.5, 1.5) #Esto es decorativo
#plt.yticks(()) #Esto es decorativo



plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt



n = 12

X = np.arange(n)#Esto generara un array de 12 numeros

Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) #Esto generara un numero Y no tan random

Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) #Esto generara un numero Y no tan random



#plt.axes([0.025, 0.025, 0.95, 0.95]) #Esto es decorativo

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')#Aqui se expresa x,y 
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')#Aqui se expresa x,y pero y en negativo


for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.02, round(y,2), ha='center', va= 'bottom')

for x, y in zip(X, Y2):
    plt.text(x + 0.4, -y - 0.02, round(y,2), ha='center', va= 'top')
    
#La funcion zip lo que hace es convertir los valores dentro de su parentesis en una tupla
#El objetivo de este for es definir dinamicamente las variables donde caeran los textos



#plt.xlim(-.5, n) #decorativo
#plt.xticks(()) #decorativo
#plt.ylim(-1.25, 1.25) #decorativo
#plt.yticks(()) #decorativo

plt.show()
```

## 4. A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 4 en el enunciado.

Pista: tendrás que modificar la Z.

Z = np.random.uniform(0, 1, 20)

plt.pie(Z)

```python
import numpy as np

import matplotlib.pyplot as plt


n = 20

Z = np.ones(n)

Z[-1] = Z[-1]* 2

plt.pie(Z, explode=Z*.2, colors = ['%f' % (i/n) for i in range(n)])

#explode separa cada pedazo del grafico

#El ki de este ejercicio esta en el parametro colors
#'%f'% tomara el valor de la variable mas cercana, en este caso seria el resultado de (i/float(n))
#Aqui se esta definiendo un for de una sola linea dentro de una lista para asi crear una lista
#Y asignarle un color nuevo a cada valor de la lista
#Basicamente cada valor de la lista sera: [1/20,2/20,3/20,4/20,5/20...]
#Osea: [0.05,0.1,0.15,0.20,0.25,...] en otras palabras sera de 0 acercandose a 1 
#Toma ese color gradiente porque 0 es negro y 1 es blanco, mientras mas cercano a 1 mas blanco sera
#Por ultimo, '%f'% permite tambien convertir automaticamente cada elemento a string, y por eso no da error

plt.show()
```
