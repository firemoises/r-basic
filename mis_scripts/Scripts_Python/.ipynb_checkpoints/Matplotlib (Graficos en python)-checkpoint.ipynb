## Ejercicio 7

```python
import numpy as np
```

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0.0,5.0,0.1)
x2 = np.arange(0.0,5.0,0.2)

plt.figure(1)
plt.subplot(1,2,2) 
#Subplot:Es para mostrar 2 o mas graficos en una misma ventana
#Subplot: El primer argumento es la cant de filas, y el segundo agumento es la cantidad de columnas 
#y el tercer argumento argumento es la ubicacion (se puede elegir 1 de 4) IMPORTANTE el cuadrante nunca puede ser el mismo
#Subplot:2,1,1 significa el doble de columnas (ponlo mas ancho) y 1,2,2 significa el doble de filas (ponlo mas vertical)
plt.plot(x1,f(x1),"b-",x2,f(x2),"g-") 
plt.xlabel("eje de abcisas")
plt.ylabel("eje de ordenadas")

plt.figure(2)
#Figure:Se utiliza para definir que apartir de esta funcion se esta hablando de otra figura, se usa de modo secuencial
plt.subplot(1,2,2) 
plt.plot(x2+1,f(x2),"b-",x2,f(x2),"g-") 

plt.show()
```

## Ejercicio 8 textos en los graficos

```python
x1 = np.arange(0.0,5.0,0.1)
x2 = np.arange(0.0,5.0,0.2)

plt.figure(1, figsize=(10,6),dpi=90)#figsize es para definir el tamano del grafico y dpi para definir la cant de pixeles
plt.subplot(2,1,1) 
plt.plot(x1,f(x1),"b-",x2,f(x2),"g-") 
plt.xlabel("eje de abcisas")
plt.ylabel("eje de ordenadas")
plt.title("Aqui va el titulo", fontsize="14", color="green")
#plt.text(2,0.5,'sdsdsd',fontsize="8", color="red") #Texto 
plt.text(2,0.5,r'$x^2$',fontsize="11", color="red")#(para escribir en LATEX debes poner la r antes de la comilla)
plt.grid(True) #Parrilla
plt.axis([0,6,-1,2]) 

plt.annotate("Maximo local", xy=(2.1,0.5), xytext=(2.5,1), arrowprops = dict(facecolor = "black",shrink=0.05))
# Para crear una flecha

plt.show()
```

## Ejercicio 9 Historigramas

```python
# Para hacer un histograma
mu = 100
sigma = 20

x = mu + sigma + np.random.randn(10000)
n, bins, patches = plt.hist(x,50,normed=1,facecolor="g",alpha=0.6)
#El profesor aqui esta definiendo estas variables (n, bins, patches) pero n realiidad no hace falta hacerlo
#De hecho se ve como ni siquiera las usa por lo que no son necesarias
#Pero solo para tenerlo anotado, estas variables son 3 valores que devuelve plt.hist al invocarse
#Por ultimo normed=1 significa que se estan convirtiendo los valores del grafico para que vayan de 0 a 1
#Porque simplemente se ve mejor asi

plt.subplot(1,1,1) 
plt.xlabel("eje de abcisas")
plt.ylabel("eje de ordenadas")
plt.title("Aqui va el titulo")
plt.show()
```

## Ejercicio 10 Cambios de las escalas

```python
# Para hacer un histograma
mu = 0.5
sd = 0.3

y = mu + sd * np.random.randn(1000)

y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))

plt.figure(1,figsize=(10,6))
plt.subplot(2,2,1) 
plt.plot(x,y)
plt.yscale("linear") #Esto es para representar el eje y en escala linear(La escala por default, osea no es necesario)
plt.xscale("linear") #Esto es para representar el eje x en escala linear(La escala por default, osea no es necesario)
plt.title("Escala lineal")
plt.grid(True)


plt.subplot(2,2,2) #Veras el segundo grafico mas abajo a la izquierda ya que demuestra que esta en el 2do cuadrante
plt.plot(x,y)
plt.yscale("log")  #Esto es para representar el eje y en escala logaritmica
plt.title("Escala log") 
plt.grid(True)


plt.subplot(2,2,4) #Veras el segundo grafico mas abajo a la izquierda ya que demuestra que esta en el 4to cuadrante
plt.plot(x,y)
plt.yscale("logit")  #Esto es para representar el eje y en escala logit
plt.title("Escala logit") 
plt.grid(True)
from matplotlib.ticker import NullFormatter #Esto es para importar el NullFormatter (Usado para anular formatos)
cuarto = plt.gca() #Para obtener el eje actual, osea el grafico que esta en el 4to cuadrante
cuarto.yaxis.set_minor_formatter(NullFormatter())#Para anular el formato que se ve mal del cuarto cuadrante

plt.subplots_adjust(top=0.96,hspace=0.30) #Para ajustar el tamao y margenes de los graficos

plt.show()
```

## Ejercicio 11 Cambios de los ejes

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
#Para hacer un arreglo de -3.14 a 3.14 con 256 elementos. El endpoint es para indicarle que queremos incluir el ultimo digito
s, c = np.sin(x),np.cos(x) #Se pueden declaran 2 variables de golpe, separandolas con una coma
```

```python
plt.figure(1,figsize=(10,6))
plt.plot(x,s, color="blue", linewidth="1.2",linestyle="--", label="Seno")
plt.plot(x,c, color="red", linewidth="1.2",linestyle="-", label="Coseno")
plt.xlim(x.min()*1.1,x.max()*1.1)#Aqui estoy generando un eje x dinamico segun el array que esta en la eje x en ambos graficos
plt.ylim(-1.2,1.2)
plt.grid(True)
plt.xticks(np.linspace(-4,4,9)) #Para decirle especificamente cuantos puntos quiero en el eje x

ax = plt.gca() #Para traerme el eje actual
ax.spines['right'].set_color('none') #Para colorear el eje derecho pero en este caso lo elimine
ax.spines['top'].set_color('none') 
ax.spines['bottom'].set_position(['data',0])#Para centrar el eje x
ax.spines['left'].set_position(['data',0])#Para centrar el eje y

x0 = 2*np.pi/3
plt.plot([x0,x0],[0,np.sin(x0)], color="purple", linestyle="--",label="Linea extra")
plt.scatter([x0,],[np.sin(x0),], color="orange",label="Punto")
#Para hacer un punto, Mira como solo hace falta indicar una coordenada

plt.annotate("Punto", xy=(x0,np.sin(x0)), xytext=(+0,+10), textcoords = "offset points")
#textcoords permite decirle que deseas definir las coordenadas a partir de una referencia

plt.legend(loc="upper left") #Para agregar una leyenda, ten en cuenta que en el plot debe estar definido el label


for label in ax.get_xticklabels() + ax.get_yticklabels(): #Esto permite hacer un for en el conjunto de labels de los ejes
    label.set_fontsize(16) #Para agrandar el tamao de los ejes
    label.set_bbox(dict(facecolor='white',edgecolor='None',alpha=0.1))
    #facecolor es para el fondo, edgecolor es el borde de los labels y alpha es la opacidad para que no tape las lineas


plt.show()
```
